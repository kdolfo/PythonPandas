# Pandas, lendo e escrevendo


## Importando Pandas

```python
import pandas as pd
```

## Criando dados

```python
#DataFrame: dataframe é uma tabela
pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
```

```python
# DataFrame com Strings
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

```python
# Adicionando índices com index
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'],
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

```python
#Series: Em contraste é uma sequência de valores. Se o DatraFrame é uma tabela, a Series é uma lista
pd.Series([1, 2, 3, 4, 5])
```

```python
#Series com índices, funciona semelhante ao DataFrame
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

## Indexação em Pandas

### Seleção baseada em índice

```python
#Seleciona os dados baseado em sua posição numérica
reviews.iloc[0]
```

```python
#Como trazer no modelo coluna-linha
reviews.iloc[:, 0]
```

```python
#Para trazer os 3 primeiros
reviews.iloc[:3, 0]
```

```python
#Para trazer somente o segundo e terceiros itens
reviews.iloc[1:3, 0]
```

```python
#Também é possível passar uma lista
reviews.iloc[[0, 1, 2], 0]
```

```python
#É possível utilizar números negativos, assim como nas listas
reviews.iloc[-5:]
```

### Seleção baseada em "label"

```python
#Nesse paradigma é o valor do índice, e não a sua posição o quê importa
reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

### Manipulando o índice

```python
reviews.set_index("title")
```

### Seleção Condicional

```python
#Checando se são italianos
reviews.country == 'Italy'
```

```python
#Separando os italianos do resto
reviews.loc[reviews.country == 'Italy']
```

```python
#Como tem muitos vamos filtrar os com pontuação acima de 90
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
#Ou Italiano ou >= 90
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```
