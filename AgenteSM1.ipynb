```python
#imports de bibliotecas
import pandas as pd
import plotly.express as px
```

```python
df_cursos = pd.read_csv('../content/drive/MyDrive/MachineLearning/censo_ed_sup_2023/MICRODADOS_CADASTRO_CURSOS_2023.csv',encoding="utf-8", sep="," , low_memory=False)
```

```python
print(df_cursos.columns.tolist())  # veja se SG_UF_IES está direitinho
```

```python
print('NO_CURSO' in df_cursos)
```

```python
#Verificando quais cursos existem na tabela.
print(df_cursos['NO_CURSO'].dropna().unique()[:50])
```

```python
#Filtre os cursos presenciais da área "Ciências Humanas" em instituições públicas federais.
#Como não existe escolhi outro curso
df_no_curso = df_cursos[(df_cursos['NO_CURSO'] == 'Ciência E Tecnologia') & (df_cursos['TP_CATEGORIA_ADMINISTRATIVA'] == 1)]
print(df_no_curso[['NO_CURSO','TP_CATEGORIA_ADMINISTRATIVA']])
```

```python
print(df_cursos['SG_UF'].dropna().unique()[:50])
```

```python
#Gere uma tabela com o número de cursos por UF, ordenada do maior para o menor.
df_cursos_por_uf = df_cursos.groupby('SG_UF')['NO_CURSO'].count().sort_values(ascending=False)
print(df_cursos_por_uf)
```

```python
#Crie um dataframe com a modalidade de ensino mapeada por nome (ex: 1 = Presencial, 2 = EAD).

#Dicionário de modalidade:
modalidade = {1: 'Presencial', 2: 'EAD'}

df_cursos['TP_MODALIDADE_ENSINO'] = pd.to_numeric(df_cursos['TP_MODALIDADE_ENSINO'], errors='coerce')
df_cursos['MODALIDADE'] = df_cursos['TP_MODALIDADE_ENSINO'] \
.map(modalidade)
df_cursos['MODALIDADE'] = df_cursos['MODALIDADE'].fillna('Desconhecida')

df_cursos.to_csv('cursos_por_uf.csv', index=False)
```

```python
df_cursos[['NO_CURSO', 'SG_UF', 'MODALIDADE']]
```

```python
print(df_cursos['MODALIDADE'].dropna().unique()[:50])
```

```python
#Checando os cursos
num_presencial = (df_cursos['MODALIDADE'] == 'Presencial').sum()
print(f"Tem {num_presencial} cursos presenciais.")
num_ead = (df_cursos['MODALIDADE'] == 'EAD').sum()
print(f"Tem {num_ead} cursos EAD.")
```

```python
#Faça um gráfico de barras horizontal com os 10 estados com mais cursos da área de negócios.
#Não tem então faço com Turismo
df_cursos_turismo = df_cursos[(df_cursos['NO_CURSO'] == 'Turismo')]
df_estados_mais_cursos_turismo = df_cursos_turismo.groupby('SG_UF')['NO_CURSO'].count().sort_values(ascending=False)[:10]
df_estados_mais_cursos_turismo
```

```python
#Gerar Gráfico com os dados de cursos na área de turismo
fig = px.bar(df_estados_mais_cursos_turismo, x=df_estados_mais_cursos_turismo.values, y=df_estados_mais_cursos_turismo.index, orientation='h', title='Estados com mais cursos de turismo')
fig.show()
```

```python
#Monte um gráfico de pizza com a distribuição de cursos por categoria administrativa.

#Mapear as categorias administrativas
categoria_administrativa = {
    1: 'Pública Federal',
    2: 'Pública Estadual',
    3: 'Pública Municipal',
    4: 'Privada com fins lucrativos',
    5: 'Privada sem fins lucrativos',
    6: 'Privada - Particular em sentido estrito',
    7: 'Especial',
    8: 'Privada comunitária',
    9: 'Privada confessional'}
df_cursos['CATEGORIA_ADM'] = df_cursos.TP_CATEGORIA_ADMINISTRATIVA.map(categoria_administrativa)

# Verificando quais são as categorias administrativas existentes
#print(df_cursos['CATEGORIA_ADM'].dropna().unique()[:50])

df_cursos_por_categoria = df_cursos.groupby('CATEGORIA_ADM')['NO_CURSO'].count().sort_values(ascending=False)
df_cursos_por_categoria
```

```python
#Gerar gráfico pizza com os dados de cursos por categoria
fig = px.pie(df_cursos_por_categoria, values=df_cursos_por_categoria.values, names=df_cursos_por_categoria.index, title='Cursos por categoria administrativa')
fig.show()
```

```python
#Crie um gráfico de dispersão entre número de cursos e número de vagas ofertadas por UF.
df_cursos['QT_CURSO'] = pd.to_numeric(df_cursos['QT_CURSO'], errors='coerce')
df_cursos['QT_VG_TOTAL'] = pd.to_numeric(df_cursos['QT_VG_TOTAL'], errors='coerce')
df_cursos_vagas = df_cursos.groupby('SG_UF')[['QT_CURSO','QT_VG_TOTAL']].sum().reset_index()
df_cursos_vagas
```

```python
#gráfico de dispersão com legenda
fig = px.scatter(df_cursos_vagas, x='QT_CURSO', y='QT_VG_TOTAL', color='SG_UF', title='Número de cursos x número de vagas ofertadas por UF')
fig.show()
```

```python
#Exportar gráfico para html
fig.write_html('grafico_dispersao.html')
```
