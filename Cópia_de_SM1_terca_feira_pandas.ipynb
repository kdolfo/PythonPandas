## Criando dados

## IndexaÃ§Ã£o em Pandas

### SeleÃ§Ã£o baseada em Ã­ndice

```python
#Seleciona os dados baseado em sua posiÃ§Ã£o numÃ©rica
reviews.iloc[0]
```

```python
#Como trazer no modelo coluna-linha
reviews.iloc[:, 0]
```

```python
#Para trazer os 3 primeiros
reviews.iloc[:3, 0]
```

```python
#Para trazer somente o segundo e terceiros itens
reviews.iloc[1:3, 0]
```

```python
#TambÃ©m Ã© possÃ­vel passar uma lista
reviews.iloc[[0, 1, 2], 0]
```

```python
#Ã possÃ­vel utilizar nÃºmeros negativos, assim como nas listas
reviews.iloc[-5:]
```

### SeleÃ§Ã£o baseada em "label"

```python
#Nesse paradigma Ã© o valor do Ã­ndice, e nÃ£o a sua posiÃ§Ã£o o quÃª importa
reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

### Manipulando o Ã­ndice

```python
reviews.set_index("title")
```

### SeleÃ§Ã£o Condicional

```python
#Checando se sÃ£o italianos
reviews.country == 'Italy'
```

```python
#Separando os italianos do resto
reviews.loc[reviews.country == 'Italy']
```

```python
#Como tem muitos vamos filtrar os com pontuaÃ§Ã£o acima de 90
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
#Ou Italiano ou >= 90
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```
