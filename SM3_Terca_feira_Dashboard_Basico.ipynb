```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../content/drive/MyDrive/MachineLearning/censo_ed_sup_2023/MICRODADOS_CADASTRO_CURSOS_2023.CSV', encoding="cp1252", sep="," , low_memory=False)
#Verificando as colunas
#df_cursos.columns.tolist()
df.columns = df.columns.str.strip().str.upper()
df.head(5).to_dict()
```

```python
modalidade_map = {1: 'Presencial', 2: 'EaD'}
categoria_map = {
    1: 'Pública Federal', 2: 'Pública Estadual', 3: 'Pública Municipal',
    4: 'Privada com fins lucrativos', 5: 'Privada sem fins lucrativos',
    6: 'Privada - Particular em sentido estrito', 7: 'Especial',
    8: 'Privada comunitária', 9: 'Privada confessional'
}


# padroniza colunas
df.columns = df.columns.str.strip().str.upper()

# Criação de colunas legíveis
df['MODALIDADE'] = df['TP_MODALIDADE_ENSINO'].map(modalidade_map)
df['CATEGORIA ADMINISTRATIVA'] = df['TP_CATEGORIA_ADMINISTRATIVA'].map(categoria_map)
df['ÁREA DO CONHECIMENTO'] = df['NO_CINE_AREA_GERAL']
df['UF'] = df['SG_UF']

# Filtros que você pode alterar manualmente
filtro_modalidade = 'Presencial'
filtro_area = 'Negócios'
filtro_categoria = 'Privada com fins lucrativos'

df_filtrado = df[
    (df['MODALIDADE'] == filtro_modalidade) &
    (df['ÁREA DO CONHECIMENTO'] == filtro_area) &
    (df['CATEGORIA ADMINISTRATIVA'] == filtro_categoria)
]

# 1) Cursos por UF
px.bar(
    df_filtrado.groupby('UF').size().reset_index(name='QUANTIDADE'),
    x='UF', y='QUANTIDADE',
    title='Cursos por UF'
).show()

# 2) Distribuição por modalidade
px.pie(
    df_filtrado,
    names='MODALIDADE',
    title='Distribuição por Modalidade'
).show()

# 3) Cursos por categoria administrativa
cnt = df_filtrado['CATEGORIA ADMINISTRATIVA'] \
         .value_counts() \
         .reset_index()
cnt.columns = ['CATEGORIA ADMINISTRATIVA','QUANTIDADE']

px.bar(
    cnt,
    x='CATEGORIA ADMINISTRATIVA',
    y='QUANTIDADE',
    title='Cursos por Categoria Administrativa'
).show()

```
